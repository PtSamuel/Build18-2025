```python
import folium

# Example: Sequence of locations as (latitude, longitude)
locations = [
    (37.7749, -122.4194),  # San Francisco
    (34.0522, -118.2437),  # Los Angeles
    (36.1699, -115.1398)   # Las Vegas
]

# Initialize a map centered at the first location
map_center = locations[0]
m = folium.Map(location=map_center, zoom_start=6)

# Add markers for each location
for loc in locations:
    folium.Marker(loc).add_to(m)

# Add a line connecting the locations
folium.PolyLine(locations, color="blue", weight=2.5, opacity=1).add_to(m)

# Save the map as an HTML file
m.save("trajectory_map.html")
```

```python
from keplergl import KeplerGl

import pandas as pd

# Example data: Sequence of locations with latitude, longitude, and optional altitude
data = {
    "latitude": [37.7749, 34.0522, 36.1699],  # Latitudes
    "longitude": [-122.4194, -118.2437, -115.1398],  # Longitudes
    "altitude": [0, 0, 0]  # Optional: Altitude in meters
}
df = pd.DataFrame(data)

# Initialize Kepler.gl map
map_3d = KeplerGl(height=600)
map_3d.add_data(data=df, name="Trajectory")

# Save or display
map_3d.save_to_html(file_name="3d_trajectory_map.html")
map_3d
```

```python
import pydeck as pdk

# Example data: Sequence of locations
data = [
    {"name": "San Francisco", "lat": 37.7749, "lon": -122.4194, "elevation": 100},
    {"name": "Los Angeles", "lat": 34.0522, "lon": -118.2437, "elevation": 200},
    {"name": "Las Vegas", "lat": 36.1699, "lon": -115.1398, "elevation": 150},
]

# Create a dataframe
import pandas as pd
df = pd.DataFrame(data)

# Create a 3D arc layer
layer = pdk.Layer(
    "LineLayer",
    df,
    get_source_position=["lon", "lat"],
    get_target_position=["lon", "lat"],
    get_color=[255, 0, 0],
    get_width=5,
    auto_highlight=True,
    pickable=True,
)

# Configure the view
view_state = pdk.ViewState(
    latitude=36.0, longitude=-118.0, zoom=5, pitch=50, bearing=0
)

# Render the map
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html("3d_pydeck_map.html")
```

```python
import pydeck as pdk
import pandas as pd

# Example data: Sequence of locations forming a trajectory
data = [
    {"lat": 37.7649, "lon": -122.4194},  # San Francisco
    {"lat": 37.7749, "lon": -122.4294},  # Los Angeles
    {"lat": 37.7849, "lon": -122.4394},  # Las Vegas
]

# Create a dataframe for the trajectory
trajectory_df = pd.DataFrame({
    "path": [[(point["lon"], point["lat"]) for point in data]],  # Define path as a list of points
    "elevation": [100],  # Optional: Elevation for 3D visualization
})

# Create a PathLayer to draw the trajectory
layer = pdk.Layer(
    "PathLayer",
    trajectory_df,
    get_path="path",
    get_color=[255, 0, 0],
    width_scale=20,
    width_min_pixels=5,
    get_width=5,
)

# Configure the view
view_state = pdk.ViewState(
    latitude=36.0,  # Center of the view
    longitude=-118.0,
    zoom=5,
    pitch=50,  # Tilt for a 3D view
    bearing=0,
)

# Render the map with the trajectory
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html("3d_pydeck_trajectory_map.html")
```
